```
import numpy as np
import skimage.io
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
from google.colab import drive
drive.mount('/content/drive')
```

```
#Loading the image 
test_image = skimage.io.imread('/content/drive/My Drive/Colab Notebooks/image.jpg')

#Converting to grayscale
test_image_gray = cv2.cvtColor(test_image, cv2.COLOR_BGR2GRAY)

# Displaying the grayscale image
plt.imshow(test_image_gray, cmap='gray')
```

```
haar_cascade_face = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
```

```
faces_rects = haar_cascade_face.detectMultiScale(test_image_gray, scaleFactor = 1.2, minNeighbors = 5);

# Let us print the no. of faces found
print('Faces found: ', len(faces_rects))
```

```
for (x,y,w,h) in faces_rects:
     cv2.rectangle(test_image, (x, y), (x+w, y+h), (0, 255, 0), 2)
```

```
#convert image to RGB and show image
coloredImage = cv2.cvtColor(test_image, cv2.COLOR_BGR2RGB)
plt.imshow(coloredImage)
```

```
#write the result image in the folder you want.
cv2.imwrite('/content/drive/My Drive/Colab Notebooks/image_result.jpg', coloredImage)
```
